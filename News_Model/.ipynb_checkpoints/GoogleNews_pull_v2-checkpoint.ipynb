```python
import requests
from bs4 import BeautifulSoup
import re
import numpy as np
import pandas as pd

#What time of day?
#For Each team?


base_url = 'https://www.google.com/search?q=Toronto+Maple+Leafs&rlz=1C1CHBF_enCA869CA869&tbas=0&tbs=qdr:m,sbd:1&tbm=nws&sxsrf=ALeKk011NCFMqo0D2iLfnwF2SKLsZOGhNg:1596551924534&source=lnt&sa=X&ved=0ahUKEwiCtc6y44HrAhWlmOAKHXcbBt8QpwUIJA&biw=1366&bih=635&dpr=1'

pages = []

for t in np.arange(0,30,10):
    
    link_1 = 'https://www.google.com/search?q=Toronto+Maple+Leafs&rlz=1C1CHBF_enCA869CA869&tbas=0&tbs=qdr:m&tbm=nws&sxsrf=ALeKk02mp7OwdMzN2TtCtaDCiRCq-eoxHw:1596545930274&ei=ilspX7usENLC1QHQz5f4Cw&start='
    
    pag = str(t)
    
    link_2 = '&sa=N&ved=0ahUKEwi7maqIzYHrAhVSYTUKHdDnBb84RhDy0wMIgAE&biw=1366&bih=635&dpr=1'
    
    
    pag_lnk = link_1 + pag + link_2
    
    print(pag_lnk)
    pages.append(pag_lnk)
    
#print(len(pages))
```

```python
#url_1 = 'https://www.google.com/search?q=Toronto+Maple+Leafs&rlz=1C1CHBF_enCA869CA869&tbas=0&tbs=qdr:m&tbm=nws&sxsrf=ALeKk02mp7OwdMzN2TtCtaDCiRCq-eoxHw:1596545930274&ei=ilspX7usENLC1QHQz5f4Cw&start=0&sa=N&ved=0ahUKEwi7maqIzYHrAhVSYTUKHdDnBb84RhDy0wMIgAE&biw=1366&bih=635&dpr=1'

import dehumanize

cnt = 0 

for p in pages:

    #Pull the google topic page
    res = requests.get(p)
    html_page = res.content
    soup = BeautifulSoup(html_page, 'html.parser')

    data_links = []
    
    #Pull the urls for each page
    for link in soup.findAll('a', href = True):
        link_text = link.get('href')
        if 'url?q=' in link_text:
            comp_link = link_text[7:]
            data_links.append(comp_link)
            #cnt = cnt+1
            
    #Get the times
    art_times = []

    for time in soup.findAll('span'):
        if 'ago' in time.get_text():
            
            #print(time.get_text())
            #print(dehumanize.naturaltime(time.get_text()))
            
            art_times.append(dehumanize.naturaltime(time.get_text()))
            
    #Select useful links
    T = [data_links[i] for i in np.arange(1,len(data_links)-1,2).tolist()]
    
    #Zip times and links
    data_tuples = list(zip(art_times,T))

    #Place into a dataframe
    
    #print(cnt)
    
    if cnt == 0:
        link_time_df = pd.DataFrame(data_tuples,columns=['time','link'])
        #print(link_time_df)
    
    else:
        link_time_df_new = pd.DataFrame(data_tuples,columns=['time','link'])
        link_time_df = pd.concat([link_time_df, link_time_df_new],axis=0,ignore_index=True)
        
    cnt = cnt + 1
        
print(link_time_df.shape)
```

```python
link_time_df['time'] = pd.to_datetime(link_time_df['time'], format='%Y-%m-d H:M:S', errors='ignore') 

link_time_df
```

```python
link_text = []
parse_len = []
parse_time = []

import article_parsers
cnt = 0

for link in link_time_df['link']:
    #print(link)

    url = link.split('&')[0]
    
    res = requests.get(url)
    html_page = res.content
    soup = BeautifulSoup(html_page, 'html.parser')
    text = soup.find_all(text=True)
    
    #What is base site... 
    base_site = url.split('/')[2]
    
    final = article_parsers.base_parser(text,base_site)
    
    time = article_parsers.time_parser(soup,base_site)
    
    print([url[:25], time])        
    
    if type(time) != datetime:
        time = link_time_df.iloc[cnt].time
    
    link_text.append(final)
    parse_time.append(time)
    #Text parser for each top site... 
    #print(base_site, len(final))
    parse_len.append([url,len(final)])
    
    cnt = cnt+1

link_time_df['link_text'] = link_text

link_time_df['link_time_parse'] = parse_time

link_time_df_2 = link_time_df
```
