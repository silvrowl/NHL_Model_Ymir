```python
import requests
from bs4 import BeautifulSoup
import re

#selenium

url = 'https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNR28yZEhJU0JXVnVMVWRDS0FBUAE?hl=en-CA&gl=CA&ceid=CA%3Aen'

#url = 'https://www.google.com/search?q=Toronto+Maple+Leafs&rlz=1C1CHBF_enCA869CA869&tbas=0&tbs=qdr:m&tbm=nws&sxsrf=ALeKk007P_x37leqUaqcjzQP8BnQtLiONA:1594758881067&ei=4RYOX_TRA83s_Qb98KKoCA&start=70&sa=N&ved=0ahUKEwi0xu7ky83qAhVNdt8KHX24CIU4ChDy0wMIeA&biw=1366&bih=635&dpr=1'

res = requests.get(url)
html_page = res.content
soup = BeautifulSoup(html_page, 'html.parser')

cnt = 0 

data_links = []

for link in soup.findAll('a', attrs={'href': re.compile("^./articles")}):
    
    comp_link = 'https://news.google.com' + link.get('href')[1:]
    
    #print(comp_link)
    
    data_links.append(comp_link)
    cnt = cnt+1
print(soup)
```

```python
data_times = []

cnt = 0 
for link in soup.findAll('time'):
    link_dt = link.get('datetime')
    
    data_times.append(link_dt)
    
    cnt=cnt+1
```

```python
len(data_links)

mylist = list(dict.fromkeys(data_links))

data_links = mylist
```

```python
import pandas as pd

data_tuples = list(zip(data_times[0:50],data_links[0:50]))

link_time_df = pd.DataFrame(data_tuples,columns=['time','link'])
```

```python
link_text = []


for link in link_time_df['link']:
    #print(link)

    url = link
    
    res = requests.get(url)
    html_page = res.content
    soup = BeautifulSoup(html_page, 'html.parser')
    text = soup.find_all(text=True)

    output = ''
    blacklist = [
        '[document]',
         'a',
         'aside',
         'article'
         'b',
         'body',
         'div',
         'fieldset',
         'figure',
         'footer',
         'form',
         'h1',
         'h2',
         'h6',
         'head',
         'header',
         'html',
         'input',
         'li',
         'link',
         'meta',
         'noscript',
         'script',
         'section',
         'span',
         'strong',
         'style',
         'time',
         'title',
         'ul']

    for t in text:
        if t.parent.name not in blacklist:
            output += '{} '.format(t)
            
    final = output.replace('\n','')
    
    link_text.append(final)

data_tuples_2 = list(zip(data_times[0:50],data_links[0:50],link_text))

link_time_df_2 = pd.DataFrame(data_tuples_2,columns=['time','link','text'])
```

```python
link_time_df_2
```

```python
#Now the sentiment analysis....
import nltk
nltk.download('vader_lexicon')

from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyser = SentimentIntensityAnalyzer()


sent_list = []

for article in link_time_df_2['text']:
    
    scores = analyser.polarity_scores(article)
    sent_list.append(scores)
```

```python
sent_df = pd.DataFrame(sent_list)
sent_df
```

```python
all_data = pd.concat([link_time_df_2,sent_df],ignore_index=True,axis=1)
all_data.columns = ['time','link','text','neg','neu','pos','compound']
all_data
```
